```python
from genshintoolkit.gtcommon.atfdata import *
from genshintoolkit.gtcommon.gtenum import *

bow_baseatk=542
bow_cr=0.441
bow_dmgb=0.12
bow_basedmgbonus=lambda em:em*1.6

bow_4star_baseatk=454
bow_4star_atkp=0.551
bow_4star_em=140

tighnari_baseatk=268
tighnari_dmgb_1=0.288
tighnari_em=50
tighnari_dmgb_a=lambda x : x*0.0008

dendro_reasonate_em=80

multplier_charged=1.5696
multplier_split=0.6948

multiplier_e=2.6928

multiplier_q_main=1.0012
multiplier_q_split=1.2236

bennett_q=lambda x : x*1.39
collei_em=60

enemy_res_factor=1.2

atf_em_2pcs=80
atf_em_4pcs_atkp=0.14*1
atf_em_4pcs_em=50*2
nob_atkp=0.2
instr_em=120

multiplier_cat=[0 for i in range(100)]
multiplier_cat[80]=1077
multiplier_cat[90]=1447*1.15
reac_cat_basedmg=lambda x : multiplier_cat[90]*(1+5*x/(x+1200))
```

```python
def calc_tignari_dmg(atk_percent,em,cr,cd):
    #base dmg
    atk=(tighnari_baseatk+bow_baseatk)*(1+atk_percent)+bennett_q(799)+311
    charged_main=atk*multplier_charged+bow_basedmgbonus(em)
    charged_split=atk*multplier_split+bow_basedmgbonus(em)
    sum_e=((tighnari_baseatk+bow_baseatk)*(1+atk_percent)*2+311*2+bennett_q(799))*(1+0.466+bow_dmgb+tighnari_dmgb_1)*(1+(0.05+cr)*(0.5+cd))*enemy_res_factor*0.5
    q_main=atk*multiplier_q_main
    q_split=atk*multiplier_q_split
    
    common_factor=(1+0.466+tighnari_dmgb_1+tighnari_dmgb_a(em)+bow_dmgb)*(1+(0.05+cr)*(0.5+cd))*enemy_res_factor*0.5

    sum_charged=(charged_main+4*charged_split)*6*common_factor
    sum_q=(q_main+q_split)*6*common_factor

    sum_all=sum_e+sum_charged+sum_q

    
    print(f'Tighnari cause {round(sum_all)} damage in total')
    print(f'Charged: {round(sum_charged)} or {sum_charged/sum_all:0.0%}')
    print(f'e: {round(sum_e)} or {sum_e/sum_all:0.0%}')
    print(f'Q: {round(sum_q)} or {sum_q/sum_all:0.0%}')
    print(f'Realtime stats:')
    print(f'ATK:{round(atk)}|EM:{em}|CR:{cr+0.05:0.0%}|CD:{cd+0.5:0.0%}')
    
print('Using em sands')
calc_tignari_dmg(
    nob_atkp+atf_em_4pcs_atkp,
    187+substat_values_avg[EnumStats.elemental_mastery]*6+dendro_reasonate_em+tighnari_em+atf_em_2pcs+atf_em_4pcs_em,
    10*substat_values_avg[EnumStats.crit_rate]+bow_cr,
    8*substat_values_avg[EnumStats.crit_dmg]+0.662
)
print('Using atk sands')
calc_tignari_dmg(
    nob_atkp+0.466+substat_values_avg[EnumStats.atk_percent]*6,
    dendro_reasonate_em+tighnari_em+atf_em_2pcs+atf_em_4pcs_em,
    10*substat_values_avg[EnumStats.crit_rate]+bow_cr,
    8*substat_values_avg[EnumStats.crit_dmg]+0.662
)
```

```python
def calc_tignari_dmg_4stars(atk_percent,em,cr,cd):
    #base dmg
    atk=(tighnari_baseatk+454)*(1+atk_percent)+bennett_q(799)+311
    charged_main=atk*multplier_charged
    charged_split=atk*multplier_split
    sum_e=((tighnari_baseatk+454)*(1+atk_percent)*2+311*2+bennett_q(799))*(1+0.466+tighnari_dmgb_1)*(1+(0.05+cr)*(0.5+cd))*enemy_res_factor*0.5
    q_main=atk*multiplier_q_main
    q_split=atk*multiplier_q_split
    
    common_factor=(1+(0.05+cr)*(0.5+cd))*enemy_res_factor*0.5

    sum_charged=(charged_main+4*charged_split)*6*common_factor*(1+0.466+0.35+tighnari_dmgb_1+tighnari_dmgb_a(em))
    sum_q=(q_main+q_split*4)*6*common_factor*(1+0.466+tighnari_dmgb_1+tighnari_dmgb_a(em))

    sum_all=sum_e+sum_charged+sum_q

    
    print(f'Tighnari cause {round(sum_all)} damage in total')
    print(f'Charged: {round(sum_charged)} or {sum_charged/sum_all:0.0%}')
    print(f'e: {round(sum_e)} or {sum_e/sum_all:0.0%}')
    print(f'Q: {round(sum_q)} or {sum_q/sum_all:0.0%}')
    print(f'Realtime stats:')
    print(f'ATK:{round(atk)}|EM:{em}|CR:{cr+0.05:0.0%}|CD:{cd+0.5:0.0%}')
    
print('Using em sands,4*bow,wanderer-4pcs')
calc_tignari_dmg_4stars(
    nob_atkp+0.551,
    187+substat_values_avg[EnumStats.elemental_mastery]*6+dendro_reasonate_em+tighnari_em+80+140,
    8*substat_values_avg[EnumStats.crit_rate]+0.331,
    10*substat_values_avg[EnumStats.crit_dmg]
)
```

```python
def calc_tignari_dmg_4stars_new4pc(atk_percent,em,cr,cd):
    #base dmg
    atk=(tighnari_baseatk+bow_4star_baseatk)*(1+atk_percent)+311
    charged_main=atk*multplier_charged
    charged_split=atk*multplier_split
    sum_e=((tighnari_baseatk+bow_4star_baseatk)*(1+atk_percent)*2+311*2+2*reac_cat_basedmg(em))\
        *(1+0.466+tighnari_dmgb_1)*(1+(0.05+cr)*(0.5+cd))*enemy_res_factor*0.5
    q_main=atk*multiplier_q_main
    q_split=atk*multiplier_q_split
    
    common_factor=(1+(0.05+cr)*(0.5+cd))*enemy_res_factor*0.5*(1+0.466+tighnari_dmgb_1+tighnari_dmgb_a(em))

    each_charged=(charged_main+4*charged_split+3*reac_cat_basedmg(em))*common_factor #主箭头触发一次，分裂箭触发两次
    each_q=((q_main+q_split)+(2/3)*reac_cat_basedmg(em))*common_factor #六根主箭触发2次，六根副剑触发2次
    sum_all=sum_e+each_charged*6+each_q*6

    print(f'提纳里在一个循环内造成 {round(sum_all)} 的总伤害, 20秒DPS为{round(sum_all/20)}')
    print('手法钟长e 提e 科莱/草主eq 雷e 提ZZZQEZZZ')
    print(f'重击: {round(each_charged*6)} or {each_charged*6/sum_all:0.0%}, {round(each_charged):0.0f} each')
    print(f'E: {round(sum_e)} or {sum_e/sum_all:0.0%}')
    print(f'Q: {round(each_q*6)} or {each_q*6/sum_all:0.0%}, avg {round(each_q):0.0f} each')
    print(f'实际面板:')
    print(f'ATK:{round(atk)}|EM:{em}|CR:{cr+0.05:0.0%}|CD:{cd+0.5:0.0%}')
    
print('Using em sands,4*bow,new-4pcs')
calc_tignari_dmg_4stars_new4pc(
    bow_4star_atkp+atf_em_4pcs_atkp,
    187+substat_values_avg[EnumStats.elemental_mastery]*6+dendro_reasonate_em+tighnari_em+atf_em_2pcs+atf_em_4pcs_em+bow_4star_em,
    8*substat_values_avg[EnumStats.crit_rate]+0.331,
    10*substat_values_avg[EnumStats.crit_dmg]
)
```
